```python
import time
from getting_accurate_solutions import *
from random import normalvariate as random_normalvariate
from auxiliary_functions_and_classes import PartialSum, partial_sum_function
from CMA import ff, fmin


#    Решение плохообусловленных матриц с помощью CMA и частичной суммы равномерно распределенных векторов
sigma = 0.5
ftarget = 1e-6
```

```python
#_________________N=6______________________________

N=6
for N1 in range(1, 10):
    print('Число слагаемых в частичной сумме =', N1)
    data2 = PartialSum(func_partial_sum=partial_sum_function, randn=np.random.uniform, N=N1, list_args=(-3,3))
    x0_uniform = [partial_sum_function(-5.12/3, 5.12/3, N=N1, randn=np.random.uniform) for i in range(N)]
    
    start = time.time()
    fmin(ff.f_ploho_obuslovlennaya_matrica_6, x0_uniform, sigma, exact_solution=the_exact_solution_6, 
         randn=data2, ftarget=ftarget)
    print(f'Время работы = {time.time() - start}')
```
